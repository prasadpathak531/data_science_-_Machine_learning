## Data Description:
- All Indian Premier League Cricket matches between 2008 and 2016.
- This is the ball by ball data of all the IPL cricket matches for 9 Seasons.
- **The dataset contains 2 files:** deliveries.csv and matches.csv.
- **matches.csv** contains details related to the match such as location, contesting teams, umpires, results, etc.
- **deliveries.csv** is the ball-by-ball data of all the IPL matches including data of the batting team, batsman, bowler, non-striker, runs scored, etc.

## Why do we have missing values?
- **City:** Data for that match is not known
- **Umpire3:** There was no umpire3 in whole IPL matches.
- **player_dismissed:** This is deliveries data (data for each ball bowled). Not every ball player is dismissed, hence we have missing values (95%)
- **dismissal_kind:** This is deliveries data (data for each ball bowled). Not every ball player is dismissed therefore kind of dismissal is not valid, hence we have missing values (95%)
- **fielder:** This is deliveries data (data for each ball bowled). Fielder is not invlolved in every ball, hence we have missing values (96%)
- **winner:** Usually matches get cancelled due to reasons like rain etc., hence that could be the reason for the missing values in this variable.
- **player_of_match:** when matches get canelled, there is no player of the match. Hence this variable has same missing value as Winner.
- **Umpire1 and umpire2:** Cancelled matches or missing data could be reason for the missing values in these variables. Hence this has the least missing values. If match starts for even 1 ball, we need minimum 2 field umpires.

## How do we fill the missing values?
- **City:** Replace with ``INFO_MISSING``
- **winner:** Replace with ``INFO_MISSING`` 
- **player_of_match:** Replace with ``INFO_MISSING`` 
- **Umpire1 and umpire2:** Replace with ``INFO_MISSING``
- **Umpire3:** Drop the variable as there is no data in this variable.
- **player_dismissed:** Replace with ``Not Applicable``
- **dismissal_kind:** Replace with ``Not Applicable``
- **fielder:** Replace with ``Not Applicable``

```python
df.city.fillna('INFO_MISSING', inplace = True)
df.umpire1.fillna('INFO_MISSING', inplace = True)
df.umpire2.fillna('INFO_MISSING', inplace = True)
df.winner.fillna('INFO_MISSING', inplace = True)
df.player_of_match.fillna('INFO_MISSING', inplace = True)
df.drop('umpire3', axis = 1, inplace = True)


df.player_dismissed.fillna('Not Applicable', inplace = True)
df.dismissal_kind.fillna('Not Applicable', inplace = True)
df.fielder.fillna('Not Applicable', inplace = True)

df.isnull().sum()
```
