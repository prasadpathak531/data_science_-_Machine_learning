```python
data = pd.read_csv('trainn.csv')
data.head()
```

```python
X = data.drop('label', axis = 1).values
```

```python
X.shape
```

```python
plt.figure()
plt.imshow(X[5].reshape(28,28), cmap = 'gray')
plt.axis('off')
```

```python
# Printing multiple images in same figure
plt.figure(figsize = (10,10))
for i in range(400):
    plt.subplot(20,20,i + 1)
    plt.axis('off')
    plt.imshow(X[i].reshape(28,28), cmap = 'gray')
```

```python
x = np.arange(30)
y = x + np.random.randn(30)
```

```python
plt.figure()
plt.plot(x, y, 'm')
```

```python
plt.figure(dpi = 100)
plt.plot(x,y,'g--')
```

```python
plt.figure(dpi = 100)
plt.plot(x,y,'m--o')
```

```python
x = np.arange(30)
plt.figure(dpi = 100)
plt.plot(x, x**2, 'm-o', label = 'Squares')
plt.plot(x, x**3, 'c--*', label = 'Cubes')
plt.title('Squares and Cubes')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.xticks(range(0,30,3))
plt.yticks(range(0,30000,3000))
plt.legend()
```

```python
#Scatterplots
x = np.arange(50)
y = x + np.random.randn(50)
plt.figure(dpi = 100)
plt.scatter(x,y, c = 'b', s = 30)
```

```python
# Barplot
x = np.arange(10)
y = x * 2
plt.figure(dpi = 100)
plt.bar(x,y, width=0.7, color = 'g')
```

```python
# Histogram
x = np.random.randint(1,30,1000)
plt.hist(x)
```

```python
plt.figure()
plt.hist([1,2,3,4,5,4,4,5,6,7,6,5,4,4,6,5,7,5,6,4,5,6,9,0,1,8,8,3,3,3,2,2])
plt.xticks(range(10))
```

```python
# Piechart
x = np.array([10,20,30,20,30])
plt.pie(x)
```

```python
plt.figure(dpi = 100)
x = np.array([10,20,30,20,30])
fruits = ['Bananas', 'Apples', 'Mangoes','Big Bananas','Tiny Bananas']
plt.pie(x, labels = fruits)
```

```python
# Boxplot
x = np.random.randint(1,100,1000)
plt.figure(dpi = 100)
plt.boxplot(x)
```

```python
# violinplot
plt.figure()
x = np.random.randint(1,100,1000)
plt.violinplot(x)
```
