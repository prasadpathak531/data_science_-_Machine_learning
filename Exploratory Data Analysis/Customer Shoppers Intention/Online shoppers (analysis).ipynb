```python
fig, axes = plt.subplots(1, 2, figsize=(18, 7))
Browser_to_Revenue = pd.DataFrame(df, columns=['Browser', 'Revenue'])
Browser_to_Revenue.groupby('Browser').count().plot(
    kind='bar', ax=axes[0], ylabel='count', legend=False)
Browser_to_Revenue.groupby('Browser').mean().plot(
    kind='bar', ax=axes[1], ylabel='mean Revenue', legend=False)
```

```python
ExitRates_to_BounceRates = pd.DataFrame(
    df, columns=['ExitRates', 'BounceRates'])
ExitRates_to_BounceRates.ExitRates = ExitRates_to_BounceRates.ExitRates.round(
    decimals=3)
pd.concat([
    ExitRates_to_BounceRates.groupby('ExitRates').min().rename(
        columns={'BounceRates': 'minimal BounceRates'}),
    ExitRates_to_BounceRates.groupby('ExitRates').max().rename(
        columns={'BounceRates': 'maximal BounceRates'}),
    ExitRates_to_BounceRates.groupby('ExitRates').mean().rename(
        columns={'BounceRates': 'mean BounceRates'}),
], axis=1,).plot()
```
