## Selecting Features

```python
cols = ['Vehicle Class', 'Transmission', 'Fuel Type','Cylinders', 'Engine Size(L)', 'Fuel Consumption Comb (L/100 km)']
X = data[cols]
Y = data[['CO2 Emissions(g/km)']]
X.head()
```

```python
from sklearn.preprocessing import OrdinalEncoder
encoder = OrdinalEncoder()
cols = ['Vehicle Class', 'Transmission', 'Fuel Type']
X[cols] = encoder.fit_transform(X[cols])
```

```python
X.head()
```

```python
print('Independent features Shape : ', X.shape)
print('Dependent features Shape   : ', Y.shape)
```

## Splitting the Data

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size = 0.25, random_state = 5)
```

```python
print('Training Data Shape   : ', x_train.shape)
print('Training Labels Shape : ', y_train.shape)
print('Testing Data Shape    : ', x_test.shape)
print('Testing Labels Shape  : ', y_test.shape)
```

## Fitting The Model

```python
from sklearn.neighbors import KNeighborsRegressor
regressor = KNeighborsRegressor(n_neighbors = 3)
regressor.fit(x_train, y_train)
```

```python
y_pred_train = regressor.predict(x_train)
y_pred_test = regressor.predict(x_test)
```

```python
from sklearn.metrics import r2_score
print('Training Accuracy : ', r2_score(y_train, y_pred_train))
print('Testing Accuracy  : ', r2_score(y_test, y_pred_test))
```

```python
# Very low value of K (n_neighbors)
regressor = KNeighborsRegressor(n_neighbors = 1)
regressor.fit(x_train, y_train)

y_pred_train = regressor.predict(x_train)
y_pred_test = regressor.predict(x_test)

print('Training Accuracy : ', r2_score(y_train, y_pred_train))
print('Testing Accuracy  : ', r2_score(y_test, y_pred_test))
```

```python
# Very high value of K (n_neighbors)
regressor = KNeighborsRegressor(n_neighbors = 100)
regressor.fit(x_train, y_train)

y_pred_train = regressor.predict(x_train)
y_pred_test = regressor.predict(x_test)

print('Training Accuracy : ', r2_score(y_train, y_pred_train))
print('Testing Accuracy  : ', r2_score(y_test, y_pred_test))
```

```python
acc_train = []
acc_test = []

for i in range(1,51):
    regressor = KNeighborsRegressor(n_neighbors = i)
    regressor.fit(x_train, y_train)

    y_pred_train = regressor.predict(x_train)
    y_pred_test = regressor.predict(x_test)

    acc_train.append(r2_score(y_train, y_pred_train))
    acc_test.append(r2_score(y_test, y_pred_test))
```

```python
x = list(range(1,51))

plt.figure(figsize = (12,6), dpi = 300)
plt.plot(x, acc_train, '-o', label = 'Training Accuracy')
plt.plot(x, acc_test,  '-o', label = 'Testing Accuracy')

plt.xticks(range(1,51,2))
plt.xlabel('Values Of K')
plt.ylabel('Accuracies')
plt.title('Comparing results for different values of K')
plt.legend()
```
