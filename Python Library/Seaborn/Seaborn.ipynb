```python
col1 = np.linspace(0,10,1000)
col2 = np.sin(col1)
df = pd.DataFrame({'C1' : col1, 'C2' : col2})
df.head(10)
```

```python
df.info()
```

```python
plt.figure(figsize = (8,4), dpi = 200)
plt.style.use('seaborn-darkgrid')
sns.lineplot(df.C1, df.C2)
```

```python
plt.figure(figsize = (8,4), dpi = 200)
plt.style.use('seaborn-darkgrid')
sns.lineplot('C1', 'C2', data = df)
```

```python
mpl.rcParamsDefault
```

```python
plt.figure(figsize = (12,6), dpi = 200)
sns.set(rc = {'axes.facecolor': 'cyan', 'axes.grid':False, 'xtick.labelsize' : 14, 'ytick.labelsize':14,'axes.labelsize' : 16})
sns.lineplot(df.C1, df.C2, linewidth = 3)
sns.lineplot(df.C1, df.C2 + 1, linewidth = 3)
```

```python
plt.figure(figsize = (12,6), dpi = 200)
sns.lineplot(df.C1, df.C2, linewidth = 3, color = 'magenta')
sns.lineplot(df.C1, df.C2 + 1, linewidth = 3, color = 'orange')
```

```python
plt.figure(figsize = (12,6), dpi = 200)
sns.set(rc = {'axes.facecolor': '#c5e6e8', 'axes.grid':False, 'xtick.labelsize' : 14, 'ytick.labelsize':14,'axes.labelsize' : 16})
sns.lineplot(df.C1, df.C2, linewidth = 3, color = '#ed6921')
sns.lineplot(df.C1, df.C2 + 1, linewidth = 3, color = '#eb7cf7')
```

```python
plt.figure(figsize=(12,6),dpi = 200)
sns.set(rc = {'axes.facecolor' : '#283747', 'axes.grid' : False, 'xtick.labelsize' : 14, 'ytick.labelsize':14,'axes.labelsize' : 16})
sns.lineplot(x = df.C1, y = df.C2, data = df, linewidth = 3.0, color = '#FF5722')
sns.lineplot(x = df.C1, y = df.C2 + 1, data = df, linewidth = 3.0, color = '#FFEB3B')
```

```python
plt.figure(figsize=(12,5),dpi = 200)
sns.set(rc = {'axes.facecolor' : 'white', 'axes.grid' : False, 'xtick.labelsize' : 14, 'ytick.labelsize':14,'axes.labelsize' : 16})
sns.lineplot(x = df.C1, y = df.C2, data = df, linewidth = 3.0, color = '#FF5722', label = 'SinWave')
sns.lineplot(x = df.C1, y = df.C2 + 1, data = df, linewidth = 3.0, color = '#FFEB3B', label = 'SinWave + 1')
plt.legend()
plt.title('Cool stuff with seaborn')
```

```python
auto = pd.read_csv('Automobile_Data.csv')
auto.head(10)
```

```python
auto.info()
```

```python
auto.replace('?',np.nan, inplace = True)
```

```python
auto.duplicated().sum()
```

```python
auto.isnull().sum()
```

```python
auto.columns
```

```python
auto.price = auto.price.apply(float)
auto.horsepower = auto.horsepower.apply(float)
auto['peak-rpm'] = auto['peak-rpm'].apply(float)
```

```python
auto.info()
```

```python
auto.nunique()
```

```python
print('Fuel Types       : ',auto['fuel-type'].unique())
print('Body Styles      : ',auto['body-style'].unique())
print('Wheel Drives     : ',auto['drive-wheels'].unique())
print('Engine Location  : ',auto['engine-location'].unique())
print('Engine Types     : ',auto['engine-type'].unique())
print('No. of Cylinders : ',auto['num-of-cylinders'].unique())
print('Fuel System      : ',auto['fuel-system'].unique())
```
